```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree,metrics
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/NeilPandey/Test_Dataset/refs/heads/master/car.data",names=['buying','maint','doors','person'
                                                                                                               ,'lug_boot','safety','class1'])
```

```python
data['class1'],class_names=pd.factorize(data['class1'])

print(class_names)
```

```python
data['buying'],_=pd.factorize(data['buying'])

data['doors'],_=pd.factorize(data['doors'])

data['person'],_=pd.factorize(data['person'])

data['lug_boot'],_=pd.factorize(data['lug_boot'])

data['safety'],_=pd.factorize(data['safety'])

data['maint'],_=pd.factorize(data['maint'])

data.head(10)
```

```python
data.info()
```

```python
#Data prep

x=data.iloc[:,0:-1]

y=data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
dtree=tree.DecisionTreeClassifier(criterion='entropy',random_state=0)

dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)

len(y_pred)
```

```python
count_misclassified=(y_test != y_pred).sum()

print('count_misclassified samples: {}'.format(count_misclassified))
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)

print('accuracy:{:.2f}'.format(accuracy))
```
